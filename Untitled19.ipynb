```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Rectangle, FancyArrowPatch
from matplotlib.path import Path
import matplotlib.patches as patches

class OperationsEvaluation:
    def __init__(self):
        self.result = None
        self.exception_reason = None
        self.exception_note = None
        self.evaluation_path = []

    def evaluate_case(self, case):
        """
        Evaluate a case according to the operations procedure

        Parameters:
        case (dict): A dictionary containing:
            - query (str): The query text
            - answer_returned (bool): Whether an answer was returned
            - article_relevant (bool): Whether the retrieved article was correct and relevant
            - answer_complete (bool): Whether a complete answer was returned
            - is_keyword (bool): Whether the query is just keywords
            - is_out_of_scope (bool): Whether the query is out of scope

        Returns:
        dict: Evaluation results
        """
        # Check if case should be discarded
        if case.get('is_keyword', False):
            self.result = "DISCARDED"
            self.exception_reason = "Keyword query"
            self.evaluation_path = ["Start", "Discard Check", "Discarded: Keyword"]
            return self._get_result()

        if case.get('is_out_of_scope', False):
            self.result = "DISCARDED"
            self.exception_reason = "N/A query"
            self.evaluation_path = ["Start", "Discard Check", "Discarded: Out of Scope"]
            return self._get_result()

        # Begin evaluation
        self.evaluation_path = ["Start", "Discard Check", "Evaluation Start"]

        # Q1: Was an answer returned?
        self.evaluation_path.append("Q1: Answer Returned?")
        if not case.get('answer_returned', False):
            self.result = "FAIL"
            self.exception_reason = "Correctness (No Answer)"
            self.exception_note = "No query answer"
            self.evaluation_path.append("No")
            return self._get_result()

        self.evaluation_path.append("Yes")

        # Q2: Was the article retrieved correct and relevant?
        self.evaluation_path.append("Q2: Article Relevant?")
        if not case.get('article_relevant', False):
            self.result = "FAIL"
            self.exception_reason = "Context"
            self.exception_note = "Wrong and/or irrelevant article retrieved"
            self.evaluation_path.append("No")
            return self._get_result()

        self.evaluation_path.append("Yes")

        # Q3: Was a complete answer returned?
        self.evaluation_path.append("Q3: Answer Complete?")
        if not case.get('answer_complete', False):
            self.result = "FAIL"
            self.exception_reason = "Completeness"
            self.exception_note = "Answer omitted key information or was unclear"
            self.evaluation_path.append("No")
            return self._get_result()

        self.evaluation_path.append("Yes")
        self.result = "PASS"
        self.exception_reason = None
        self.exception_note = None

        return self._get_result()

    def _get_result(self):
        """Return a dictionary with the evaluation results"""
        return {
            'result': self.result,
            'exception_reason': self.exception_reason,
            'exception_note': self.exception_note,
            'evaluation_path': self.evaluation_path
        }

    def visualize_flow(self, figsize=(12, 10)):
        """Create a visualization of the operations evaluation flow"""
        fig, ax = plt.subplots(figsize=figsize)

        # Define node positions
        nodes = {
            'start': (5, 9.5),
            'discard_check': (5, 8.5),
            'discard_keyword': (2, 8),
            'discard_outofscope': (8, 8),
            'eval_start': (5, 7.5),
            'q1': (5, 6.5),
            'q1_no': (8, 6),
            'q2': (5, 5),
            'q2_no': (8, 4.5),
            'q3': (5, 3.5),
            'q3_no': (3, 2.5),
            'q3_yes': (7, 2.5)
        }

        # Draw nodes
        node_styles = {
            'start': {'color': 'royalblue', 'width': 2.5, 'height': 0.6, 'textcolor': 'white'},
            'discard_check': {'color': 'lightgrey', 'width': 2.5, 'height': 0.6, 'textcolor': 'black'},
            'discard_keyword': {'color': 'lightgrey', 'width': 2.5, 'height': 0.6, 'textcolor': 'black'},
            'discard_outofscope': {'color': 'lightgrey', 'width': 2.5, 'height': 0.6, 'textcolor': 'black'},
            'eval_start': {'color': 'royalblue', 'width': 2.5, 'height': 0.6, 'textcolor': 'white'},
            'q1': {'color': 'gold', 'width': 2.5, 'height': 0.6, 'textcolor': 'black'},
            'q1_no': {'color': 'lightcoral', 'width': 2.5, 'height': 0.8, 'textcolor': 'black'},
            'q2': {'color': 'gold', 'width': 2.5, 'height': 0.6, 'textcolor': 'black'},
            'q2_no': {'color': 'lightcoral', 'width': 2.5, 'height': 0.8, 'textcolor': 'black'},
            'q3': {'color': 'gold', 'width': 2.5, 'height': 0.6, 'textcolor': 'black'},
            'q3_no': {'color': 'lightcoral', 'width': 2.5, 'height': 0.8, 'textcolor': 'black'},
            'q3_yes': {'color': 'lightgreen', 'width': 2.5, 'height': 0.8, 'textcolor': 'black'}
        }

        # Node texts
        node_texts = {
            'start': 'Start Evaluation',
            'discard_check': 'Check for Discard Reasons',
            'discard_keyword': 'DISCARDED\nKeyword query',
            'discard_outofscope': 'DISCARDED\nN/A query',
            'eval_start': 'Begin Evaluation',
            'q1': 'Was an answer returned?',
            'q1_no': 'FAIL\nCorrectness (No Answer)\nAll remaining: N/A',
            'q2': 'Was the article retrieved\ncorrect and relevant?',
            'q2_no': 'FAIL\nContext\nAll remaining: N/A',
            'q3': 'Was a complete answer returned?',
            'q3_no': 'FAIL\nCompleteness',
            'q3_yes': 'PASS\nComplete and Correct Answer'
        }

        # Draw nodes
        for node, pos in nodes.items():
            style = node_styles[node]
            rect = Rectangle((pos[0] - style['width']/2, pos[1] - style['height']/2),
                             style['width'], style['height'],
                             facecolor=style['color'], edgecolor='black', alpha=0.8)
            ax.add_patch(rect)
            ax.text(pos[0], pos[1], node_texts[node], ha='center', va='center',
                   color=style['textcolor'], fontsize=9, fontweight='bold')

        # Define arrows
        arrows = [
            ('start', 'discard_check'),
            ('discard_check', 'discard_keyword'),
            ('discard_check', 'discard_outofscope'),
            ('discard_check', 'eval_start'),
            ('eval_start', 'q1'),
            ('q1', 'q1_no'),
            ('q1', 'q2'),
            ('q2', 'q2_no'),
            ('q2', 'q3'),
            ('q3', 'q3_no'),
            ('q3', 'q3_yes')
        ]

        # Add decision labels
        decision_labels = {
            ('discard_check', 'discard_keyword'): 'Keyword query',
            ('discard_check', 'discard_outofscope'): 'Out of scope',
            ('discard_check', 'eval_start'): 'Valid query',
            ('q1', 'q1_no'): 'No',
            ('q1', 'q2'): 'Yes',
            ('q2', 'q2_no'): 'No',
            ('q2', 'q3'): 'Yes',
            ('q3', 'q3_no'): 'No',
            ('q3', 'q3_yes'): 'Yes'
        }

        # Draw arrows
        for a, b in arrows:
            start_pos = nodes[a]
            end_pos = nodes[b]

            # Calculate arrow positions
            if a == 'q1' and b == 'q1_no':
                # Horizontal arrow
                dx = end_pos[0] - start_pos[0]
                dy = 0
                start_x = start_pos[0] + 1.25  # Move right from center of node
                start_y = start_pos[1]
                end_x = end_pos[0] - 1.25  # Move left from center of target node
                end_y = end_pos[1]
            elif a == 'q2' and b == 'q2_no':
                # Horizontal arrow
                dx = end_pos[0] - start_pos[0]
                dy = 0
                start_x = start_pos[0] + 1.25
                start_y = start_pos[1]
                end_x = end_pos[0] - 1.25
                end_y = end_pos[1]
            elif a == 'q3' and b in ['q3_no', 'q3_yes']:
                # Two arrows from q3
                dx = end_pos[0] - start_pos[0]
                dy = end_pos[1] - start_pos[1]
                start_x = start_pos[0] + (dx/abs(dx))*1.25 if dx != 0 else start_pos[0]
                start_y = start_pos[1] - 0.3
                end_x = end_pos[0]
                end_y = end_pos[1] + 0.4
            elif a == 'discard_check' and b in ['discard_keyword', 'discard_outofscope']:
                # Arrows from discard check
                dx = end_pos[0] - start_pos[0]
                dy = end_pos[1] - start_pos[1]
                start_x = start_pos[0] + (dx/abs(dx))*1.25 if dx != 0 else start_pos[0]
                start_y = start_pos[1]
                end_x = end_pos[0]
                end_y = end_pos[1] + 0.3
            else:
                # Vertical arrow
                dx = 0
                dy = end_pos[1] - start_pos[1]
                start_x = start_pos[0]
                start_y = start_pos[1] - 0.3
                end_x = end_pos[0]
                end_y = end_pos[1] + 0.3

            # Draw the arrow
            arrow = FancyArrowPatch((start_x, start_y), (end_x, end_y),
                                  arrowstyle='-|>', color='black',
                                  connectionstyle=f'arc3,rad=0.0',
                                  mutation_scale=15, linewidth=1)
            ax.add_patch(arrow)

            # Add label to arrow if applicable
            if (a, b) in decision_labels:
                # Calculate label position
                if a == 'q1' and b == 'q1_no':
                    # Label for horizontal arrow
                    label_x = start_pos[0] + (end_pos[0] - start_pos[0])/2
                    label_y = start_pos[1] + 0.2
                elif a == 'q2' and b == 'q2_no':
                    # Label for horizontal arrow
                    label_x = start_pos[0] + (end_pos[0] - start_pos[0])/2
                    label_y = start_pos[1] + 0.2
                elif a == 'q3' and b in ['q3_no', 'q3_yes']:
                    # Labels for q3 outcomes
                    label_x = start_pos[0] + (end_pos[0] - start_pos[0])/2
                    label_y = (start_pos[1] + end_pos[1])/2 + 0.1
                elif a == 'discard_check' and b in ['discard_keyword', 'discard_outofscope']:
                    # Labels for discard reasons
                    label_x = (start_pos[0] + end_pos[0])/2
                    label_y = (start_pos[1] + end_pos[1])/2 + 0.1
                else:
                    # Label for vertical arrow
                    label_x = start_pos[0] + 0.4
                    label_y = (start_pos[1] + end_pos[1])/2

                ax.text(label_x, label_y, decision_labels[(a, b)],
                       ha='center', va='center', fontsize=8)

        # Set the limits and remove axis
        ax.set_xlim(0, 10)
        ax.set_ylim(1, 10)
        ax.axis('off')

        # Add title
        ax.set_title('Operations Procedure Evaluation Flow', fontsize=14, fontweight='bold')

        # Add legend for outcome types
        legend_elements = [
            patches.Patch(facecolor='royalblue', edgecolor='black', label='Start/Process'),
            patches.Patch(facecolor='gold', edgecolor='black', label='Decision Point'),
            patches.Patch(facecolor='lightcoral', edgecolor='black', label='Fail Outcome'),
            patches.Patch(facecolor='lightgreen', edgecolor='black', label='Pass Outcome'),
            patches.Patch(facecolor='lightgrey', edgecolor='black', label='Discard Check')
        ]

        ax.legend(handles=legend_elements, loc='upper right', fontsize=8)

        plt.tight_layout()
        return fig

# Example usage
def run_example():
    evaluator = OperationsEvaluation()

    # Define test cases
    test_cases = [
        {
            'name': 'Keyword Query',
            'query': 'credit card atm',
            'is_keyword': True,
            'is_out_of_scope': False,
            'answer_returned': None,
            'article_relevant': None,
            'answer_complete': None
        },
        {
            'name': 'Out of Scope Query',
            'query': 'What is the recipe for chocolate cake?',
            'is_keyword': False,
            'is_out_of_scope': True,
            'answer_returned': None,
            'article_relevant': None,
            'answer_complete': None
        },
        {
            'name': 'No Answer Returned',
            'query': 'How do I activate my debit card?',
            'is_keyword': False,
            'is_out_of_scope': False,
            'answer_returned': False,
            'article_relevant': None,
            'answer_complete': None
        },
        {
            'name': 'Irrelevant Article',
            'query': 'How do I transfer money between accounts?',
            'is_keyword': False,
            'is_out_of_scope': False,
            'answer_returned': True,
            'article_relevant': False,
            'answer_complete': None
        },
        {
            'name': 'Incomplete Answer',
            'query': 'What are the foreign transaction fees?',
            'is_keyword': False,
            'is_out_of_scope': False,
            'answer_returned': True,
            'article_relevant': True,
            'answer_complete': False
        },
        {
            'name': 'Complete and Correct Answer',
            'query': 'How do I change my PIN?',
            'is_keyword': False,
            'is_out_of_scope': False,
            'answer_returned': True,
            'article_relevant': True,
            'answer_complete': True
        }
    ]

    # Run evaluations
    results = []
    for case in test_cases:
        results.append({
            'case': case['name'],
            'evaluation': evaluator.evaluate_case(case)
        })

    # Print results
    for result in results:
        print(f"\nCase: {result['case']}")
        print(f"Result: {result['evaluation']['result']}")
        if result['evaluation']['exception_reason']:
            print(f"Exception Reason: {result['evaluation']['exception_reason']}")
        if result['evaluation']['exception_note']:
            print(f"Exception Note: {result['evaluation']['exception_note']}")
        print(f"Evaluation Path: {' -> '.join(result['evaluation']['evaluation_path'])}")

    # Create visualization
    fig = evaluator.visualize_flow()
    plt.show()

    # Return evaluation instance for further use
    return evaluator, results

if __name__ == "__main__":
    evaluator, results = run_example()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Rectangle, FancyArrowPatch
from matplotlib.path import Path
import matplotlib.patches as patches

class ModelDeveloperEvaluation:
    def __init__(self):
        self.result = None
        self.category = None
        self.evaluation_path = []

    def evaluate_case(self, case):
        """
        Evaluate a case according to the Model Developer's method

        Parameters:
        case (dict): A dictionary containing:
            - query (str): The query text
            - is_keyword (bool): Whether the query is just keywords
            - is_in_scope (bool): Whether the query is within scope
            - retrieval_relevant (bool): Whether the retrieval is relevant
            - specific_answer_found (bool): Whether a specific answer was found in relevant content
            - answer_correct (bool): Whether the answer is correct
            - answer_complete (bool): Whether the answer is complete

        Returns:
        dict: Evaluation results
        """
        # Reset state
        self.result = None
        self.category = None
        self.evaluation_path = ["Start"]

        # Only evaluate in-scope queries
        self.evaluation_path.append("Is query in scope?")
        if not case.get('is_in_scope', True):
            self.result = "OUT OF SCOPE"
            self.category = "Out of Scope"
            self.evaluation_path.append("No")
            return self._get_result()

        self.evaluation_path.append("Yes")

        # Is this a keyword search rather than a question?
        self.evaluation_path.append("Is query a keyword search?")
        if case.get('is_keyword', False):
            self.result = "FAIL"
            self.category = "Keyword Search"
            self.evaluation_path.append("Yes")
            return self._get_result()

        self.evaluation_path.append("No")

        # Is the retrieval relevant?
        self.evaluation_path.append("Is retrieval relevant?")
        if not case.get('retrieval_relevant', False):
            self.result = "FAIL"
            self.category = "Irrelevant Retrieval"
            self.evaluation_path.append("No")
            return self._get_result()

        self.evaluation_path.append("Yes")

        # Was a specific answer found?
        self.evaluation_path.append("Specific answer found?")
        if not case.get('specific_answer_found', False):
            self.result = "FAIL"
            self.category = "Relevant Links but no specific answer"
            self.evaluation_path.append("No")
            return self._get_result()

        self.evaluation_path.append("Yes")

        # Is the answer correct?
        self.evaluation_path.append("Is answer correct?")
        if not case.get('answer_correct', False):
            self.result = "FAIL"
            self.category = "Incorrect Answer"
            self.evaluation_path.append("No")
            return self._get_result()

        self.evaluation_path.append("Yes")

        # Is the answer complete?
        self.evaluation_path.append("Is answer complete?")
        if not case.get('answer_complete', False):
            self.result = "FAIL"
            self.category = "Incomplete Answer"
            self.evaluation_path.append("No")
            return self._get_result()

        self.evaluation_path.append("Yes")
        self.result = "PASS"
        self.category = "Correct and Complete Answer"

        return self._get_result()

    def _get_result(self):
        """Return a dictionary with the evaluation results"""
        return {
            'result': self.result,
            'category': self.category,
            'evaluation_path': self.evaluation_path
        }

    def visualize_flow(self, figsize=(12, 12)):
        """Create a visualization of the Model Developer evaluation flow"""
        fig, ax = plt.subplots(figsize=figsize)

        # Define node positions
        nodes = {
            'start': (5, 10.5),
            'in_scope': (5, 9.5),
            'out_of_scope': (8, 9),
            'keyword': (5, 8.5),
            'keyword_yes': (8, 8),
            'retrieval': (5, 7.5),
            'retrieval_no': (8, 7),
            'specific_answer': (5, 6.5),
            'specific_answer_no': (8, 6),
            'answer_correct': (5, 5.5),
            'answer_correct_no': (8, 5),
            'answer_complete': (5, 4.5),
            'answer_complete_no': (8, 4),
            'pass': (5, 3.5)
        }

        # Define node styles
        node_styles = {
            'start': {'color': 'royalblue', 'width': 2.5, 'height': 0.6, 'textcolor': 'white'},
            'in_scope': {'color': 'gold', 'width': 2.5, 'height': 0.6, 'textcolor': 'black'},
            'out_of_scope': {'color': 'lightgrey', 'width': 2.5, 'height': 0.6, 'textcolor': 'black'},
            'keyword': {'color': 'gold', 'width': 2.5, 'height': 0.6, 'textcolor': 'black'},
            'keyword_yes': {'color': 'lightcoral', 'width': 2.5, 'height': 0.6, 'textcolor': 'black'},
            'retrieval': {'color': 'gold', 'width': 2.5, 'height': 0.6, 'textcolor': 'black'},
            'retrieval_no': {'color': 'lightcoral', 'width': 2.5, 'height': 0.6, 'textcolor': 'black'},
            'specific_answer': {'color': 'gold', 'width': 2.5, 'height': 0.6, 'textcolor': 'black'},
            'specific_answer_no': {'color': 'lightcoral', 'width': 2.5, 'height': 0.6, 'textcolor': 'black'},
            'answer_correct': {'color': 'gold', 'width': 2.5, 'height': 0.6, 'textcolor': 'black'},
            'answer_correct_no': {'color': 'lightcoral', 'width': 2.5, 'height': 0.6, 'textcolor': 'black'},
            'answer_complete': {'color': 'gold', 'width': 2.5, 'height': 0.6, 'textcolor': 'black'},
            'answer_complete_no': {'color': 'lightcoral', 'width': 2.5, 'height': 0.6, 'textcolor': 'black'},
            'pass': {'color': 'lightgreen', 'width': 2.5, 'height': 0.6, 'textcolor': 'black'}
        }

        # Node texts
        node_texts = {
            'start': 'Start Evaluation',
            'in_scope': 'Is query in scope?',
            'out_of_scope': 'OUT OF SCOPE\nOut of Scope',
            'keyword': 'Is query a keyword search?',
            'keyword_yes': 'FAIL\nKeyword Search',
            'retrieval': 'Is retrieval relevant?',
            'retrieval_no': 'FAIL\nIrrelevant Retrieval',
            'specific_answer': 'Specific answer found?',
            'specific_answer_no': 'FAIL\nRelevant Links but\nno specific answer',
            'answer_correct': 'Is answer correct?',
            'answer_correct_no': 'FAIL\nIncorrect Answer',
            'answer_complete': 'Is answer complete?',
            'answer_complete_no': 'FAIL\nIncomplete Answer',
            'pass': 'PASS\nCorrect and Complete Answer'
        }

        # Draw nodes
        for node, pos in nodes.items():
            style = node_styles[node]
            rect = Rectangle((pos[0] - style['width']/2, pos[1] - style['height']/2),
                             style['width'], style['height'],
                             facecolor=style['color'], edgecolor='black', alpha=0.8)
            ax.add_patch(rect)
            ax.text(pos[0], pos[1], node_texts[node], ha='center', va='center',
                   color=style['textcolor'], fontsize=9, fontweight='bold')

        # Define arrows
        arrows = [
            ('start', 'in_scope'),
            ('in_scope', 'out_of_scope'),
            ('in_scope', 'keyword'),
            ('keyword', 'keyword_yes'),
            ('keyword', 'retrieval'),
            ('retrieval', 'retrieval_no'),
            ('retrieval', 'specific_answer'),
            ('specific_answer', 'specific_answer_no'),
            ('specific_answer', 'answer_correct'),
            ('answer_correct', 'answer_correct_no'),
            ('answer_correct', 'answer_complete'),
            ('answer_complete', 'answer_complete_no'),
            ('answer_complete', 'pass')
        ]

        # Add decision labels
        decision_labels = {
            ('in_scope', 'out_of_scope'): 'No',
            ('in_scope', 'keyword'): 'Yes',
            ('keyword', 'keyword_yes'): 'Yes',
            ('keyword', 'retrieval'): 'No',
            ('retrieval', 'retrieval_no'): 'No',
            ('retrieval', 'specific_answer'): 'Yes',
            ('specific_answer', 'specific_answer_no'): 'No',
            ('specific_answer', 'answer_correct'): 'Yes',
            ('answer_correct', 'answer_correct_no'): 'No',
            ('answer_correct', 'answer_complete'): 'Yes',
            ('answer_complete', 'answer_complete_no'): 'No',
            ('answer_complete', 'pass'): 'Yes'
        }

        # Draw arrows
        for a, b in arrows:
            start_pos = nodes[a]
            end_pos = nodes[b]

            # Calculate arrow positions
            if b.endswith('_no') or b == 'out_of_scope':
                # Horizontal arrow
                dx = end_pos[0] - start_pos[0]
                dy = 0
                start_x = start_pos[0] + 1.25  # Move right from center of node
                start_y = start_pos[1]
                end_x = end_pos[0] - 1.25  # Move left from center of target node
                end_y = end_pos[1]
            else:
                # Vertical arrow
                dx = 0
                dy = end_pos[1] - start_pos[1]
                start_x = start_pos[0]
                start_y = start_pos[1] - 0.3
                end_x = end_pos[0]
                end_y = end_pos[1] + 0.3

            # Draw the arrow
            arrow = FancyArrowPatch((start_x, start_y), (end_x, end_y),
                                  arrowstyle='-|>', color='black',
                                  connectionstyle=f'arc3,rad=0.0',
                                  mutation_scale=15, linewidth=1)
            ax.add_patch(arrow)

            # Add label to arrow if applicable
            if (a, b) in decision_labels:
                # Calculate label position
                if b.endswith('_no') or b == 'out_of_scope':
                    # Label for horizontal arrow
                    label_x = start_pos[0] + (end_pos[0] - start_pos[0])/2
                    label_y = start_pos[1] + 0.2
                else:
                    # Label for vertical arrow
                    label_x = start_pos[0] + 0.4
                    label_y = (start_pos[1] + end_pos[1])/2

                ax.text(label_x, label_y, decision_labels[(a, b)],
                       ha='center', va='center', fontsize=8)

        # Set the limits and remove axis
        ax.set_xlim(0, 10)
        ax.set_ylim(2, 11)
        ax.axis('off')

        # Add title
        ax.set_title('Model Developer Evaluation Flow', fontsize=14, fontweight='bold')

        # Add legend for outcome types
        legend_elements = [
            patches.Patch(facecolor='royalblue', edgecolor='black', label='Start'),
            patches.Patch(facecolor='gold', edgecolor='black', label='Decision Point'),
            patches.Patch(facecolor='lightcoral', edgecolor='black', label='Fail Outcome'),
            patches.Patch(facecolor='lightgreen', edgecolor='black', label='Pass Outcome'),
            patches.Patch(facecolor='lightgrey', edgecolor='black', label='Out of Scope')
        ]

        ax.legend(handles=legend_elements, loc='upper right', fontsize=8)

        # Add note about LLM Faithfulness evaluation
        ax.text(5, 2.5,
                "Note: Out-of-scope queries are primarily evaluated by the LLM Faithfulness evaluator,\n"
                "which measures whether the model's responses align with the retrieved content.",
                ha='center', va='center', fontsize=8, style='italic',
                bbox=dict(facecolor='lightgrey', alpha=0.2))

        plt.tight_layout()
        return fig

# Example usage
def run_example():
    evaluator = ModelDeveloperEvaluation()

    # Define test cases
    test_cases = [
        {
            'name': 'Out of Scope Query',
            'query': 'What is the recipe for chocolate cake?',
            'is_in_scope': False,
            'is_keyword': False,
            'retrieval_relevant': None,
            'specific_answer_found': None,
            'answer_correct': None,
            'answer_complete': None
        },
        {
            'name': 'Keyword Search',
            'query': 'credit card atm',
            'is_in_scope': True,
            'is_keyword': True,
            'retrieval_relevant': None,
            'specific_answer_found': None,
            'answer_correct': None,
            'answer_complete': None
        },
        {
            'name': 'Irrelevant Retrieval',
            'query': 'How do I transfer money between accounts?',
            'is_in_scope': True,
            'is_keyword': False,
            'retrieval_relevant': False,
            'specific_answer_found': None,
            'answer_correct': None,
            'answer_complete': None
        },
        {
            'name': 'Relevant Links but no specific answer',
            'query': 'What are the foreign transaction fees?',
            'is_in_scope': True,
            'is_keyword': False,
            'retrieval_relevant': True,
            'specific_answer_found': False,
            'answer_correct': None,
            'answer_complete': None
        },
        {
            'name': 'Incorrect Answer',
            'query': 'How do I set up mobile banking?',
            'is_in_scope': True,
            'is_keyword': False,
            'retrieval_relevant': True,
            'specific_answer_found': True,
            'answer_correct': False,
            'answer_complete': None
        },
        {
            'name': 'Incomplete Answer',
            'query': 'What are the ATM withdrawal limits?',
            'is_in_scope': True,
            'is_keyword': False,
            'retrieval_relevant': True,
            'specific_answer_found': True,
            'answer_correct': True,
            'answer_complete': False
        },
        {
            'name': 'Correct and Complete Answer',
            'query': 'How do I change my PIN?',
            'is_in_scope': True,
            'is_keyword': False,
            'retrieval_relevant': True,
            'specific_answer_found': True,
            'answer_correct': True,
            'answer_complete': True
        }
    ]

    # Run evaluations
    results = []
    for case in test_cases:
        results.append({
            'case': case['name'],
            'evaluation': evaluator.evaluate_case(case)
        })

    # Print results
    for result in results:
        print(f"\nCase: {result['case']}")
        print(f"Result: {result['evaluation']['result']}")
        print(f"Category: {result['evaluation']['category']}")
        print(f"Evaluation Path: {' -> '.join(result['evaluation']['evaluation_path'])}")

    # Create visualization
    fig = evaluator.visualize_flow()
    plt.show()

    # Return evaluation instance for further use
    return evaluator, results

if __name__ == "__main__":
    evaluator, results = run_example()
```
